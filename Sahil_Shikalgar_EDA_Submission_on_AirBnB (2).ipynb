# **Project Summary -**

Exploratory Data Analysis (EDA) is a critical first step in any data science project, allowing analysts and researchers to gain a deep understanding of the dataset and its underlying patterns before diving into more complex modeling or decision-making. This project focuses on performing EDA on an Airbnb dataset to uncover key trends üìà, **outliers** üö©, **missing data** ‚ùì, and **relationships between variables** üîó‚Äîall of which can inform future actions or business decisions.

The dataset typically contains a wealth of information, such as **listings** üè°, **host details** üë§, **location** üìç, **pricing** üí∞, **availability** üìÖ, and **customer reviews** ‚≠ê. By thoroughly examining these features, we can gain valuable insights into how factors like **location** üåç, **room type** üõèÔ∏è, or the **number of reviews** üí¨ influence pricing and customer demand. Understanding these relationships can guide decision-making and help create more effective business strategies!

**Step 1: Loading and Understanding the Dataset üìä**
The first step is loading the Airbnb dataset using Python libraries like Pandas. Once loaded, functions like `info()` and `describe()` give you a quick snapshot of the data‚Äôs structure, including data types, missing values, and basic stats like mean, median, and standard deviation. This provides a first look üëÄ into how clean or messy the data is and whether it needs transformation!

**Step 2: Handling Missing and Duplicate Data üßπ**
Missing data? It happens! üïµÔ∏è‚Äç‚ôÄÔ∏è By analyzing how much data is missing in each column, you can decide to either drop or fill missing values using statistical methods (mean, median) or advanced techniques. Removing duplicate rows ensures data accuracy üßÆ, so no information is counted twice!

**Step 3: Univariate Analysis üîç**
Next, we dig deeper into individual variables. For numerical features like price or the number of reviews, we use histograms and boxplots to check their distributions. This helps spot any skewness or outliers üéØ. For categorical features like room_type and neighbourhood, count plots show the frequency of each category‚Äîhelpful for seeing which types of rooms are most common üèòÔ∏è!

**Step 4: Bivariate Analysis üîó**
Now, let‚Äôs explore the relationships between two variables at once. For example, how does the price change depending on room type or neighbourhood? üìç Boxplots can show how entire homes typically cost more than shared rooms. We can also check the correlation between numerical features like price, number_of_reviews, and availability with correlation matrices and heatmaps üî•‚Äîperfect for spotting important patterns!

**Step 5: Outlier Detection and Categorical Data Handling üö®**
Outliers can skew your analysis, especially with prices! Boxplots are a great way to identify extreme values üí°, which may need to be removed or transformed (e.g., log transformations). For categorical variables, techniques like one-hot encoding are used to convert categories into a machine-readable format‚Äîan essential step for data preprocessing üîß.

This EDA process ensures that the data is well-understood, cleaned, and ready for deeper analysis or modeling!

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Fillig the missing values,

data['name'].fillna('Unnamed_Listing', inplace = True)
data['host_name'].fillna('Unnamed_Host', inplace = True)

'''
data.fillna({name: Unnamed_Listing}, inplace=True)
data.fillna({host_name: Unnamed_Host}, inplace=True)
'''
````

The columns `name` and `host_name` were also part of the missing values count üßæ. To handle these uncertain values üåÄ, we used the `fillna` method üõ†Ô∏è to efficiently replace the missing elements. This ensures that our dataset remains complete for further analysis üîç.

Now, everything's all set for the next steps üöÄ!

````python
# calculating the percent of missing values in the last_review & reviews_per_month
missing_count_last_review = data['last_review'].isnull().sum()
total_count = len(data)

misisng_percent_last_review = (missing_count_last_review / total_count) * 100
print(f"The percentage of missing values in 'last_review' is: {misisng_percent_last_review:.2f}%")

missing_count_review_per_month = data['reviews_per_month'].isnull().sum()
total_count = len(data)
misisng_percent_reviews_per_month  = (missing_count_last_review / total_count) * 100
print(f"The percentage of missing values in 'reviews_per_month ' is: {misisng_percent_reviews_per_month :.2f}%")
````

As we can see, the DataFrame and the columns containing the missing values, such as `last_review` and `reviews_per_month`, have **20.56%** of missing values üïµÔ∏è‚Äç‚ôÄÔ∏è in both columns. Since the percentage is manageable for analysis, instead of dropping the columns üö´, we'll replace the missing values with the Median üßÆ, which is the middle value in a dataset when the values are arranged in ascending or descending order üìä.

So, let's begin filling the missing values ‚ú®....!

````python
# Convert to datetime format
data['last_review'] = pd.to_datetime(data['last_review'], errors='coerce')
data['reviews_per_month'] = pd.to_datetime(data['reviews_per_month'], errors='coerce')
````

````python
# Fill missing values with the median date
median_last_review = data['last_review'].median()
median_reviews_per_month = data['reviews_per_month'].median()

data['last_review'].fillna(median_last_review, inplace=True)
data['reviews_per_month'].fillna(median_reviews_per_month, inplace=True)
````
